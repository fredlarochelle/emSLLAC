```python
import librosa
import librosa.display
import IPython.display as ipd
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
data, sr = librosa.load('Interstellar.wav', mono=True, duration=15.0, sr=48000) 
ipd.Audio(data, rate=sr)
```

```python
def DPCM_encode(data_arr):
  differential = []

  differential.append(data_arr[0])

  for i in range(1, len(data_arr)):
    differential.append(data_arr[i] - data_arr[i-1])

  return np.array(differential)

def DPCM_decode(differential_arr):
  decoded = []

  decoded.append(differential_arr[0])
  for i in range(1, len(differential_arr)):
    decoded.append(differential_arr[i] + differential_arr[i-1])

  return np.array(decoded)
```

```python
temp = DPCM_encode(data)
test = DPCM_decode(temp)
```

```python
ipd.Audio(test, rate=sr)
```

```python
plt.figure(figsize=(28, 4))
plt.plot(data)
plt.show()

plt.figure(figsize=(28, 4))
plt.plot(temp)
plt.show()

plt.figure(figsize=(28, 4))
plt.plot(test)
plt.show()
```
